```python
import pandas as pd
# 데이터 세팅
bins = [0, 20, 30, 50, 70, 100]
labels = ['미성년자', '청년', '장년', '중년', '노년']
cats = pd.cut(titanic['age'].dropna().reset_index(drop = True), bins, labels = labels)

# 차트 속성 세팅
explode = (0.1, 0, 0, 0, 0)
colors = ["yellowgreen", "gold", "lightskyblue", "lightcoral", "orange"]
labels = cats.value_counts(normalize=True).index
sizes = cats.value_counts(normalize=True)

# 시각화
plt.title('타이타닉호 승객 나이 비율')
plt.pie(sizes, explode = explode, labels=labels, colors=colors, autopct = '%1.1f%%', shadow = True, startangle = 90)
plt.axis('equal')
plt.show()
```

```python
def f(x,y):
    return (1 - x/ 2 + x ** 5 + y ** 3) * np.exp(-x ** 2 - y ** 2)

n = 256
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)
XX, YY = np.meshgrid(x, y)
ZZ = f(XX, YY)

plt.title('Contour plots')
plt.contourf(XX, YY, ZZ, alpha = .75, cmap = 'jet')
plt.contour(XX, YY, ZZ, colors = 'black')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
XX, YY = np.meshgrid(X, Y)
RR = np.sqrt(XX**2 + YY**2)
ZZ = np.sin(RR)

fig = plt.figure()
ax = Axes3D(fig)
ax.set_title('3D surface Plot')
ax.plot_surface(XX, YY,ZZ, rstride = 1, cstride = 1, cmap = 'hot')
plt.show()
```
